# Problem  : Austro Motor Company Problem

Austo Motor Company is a leading car manufacturer specializing in SUV, Sedan, and Hatchback models. In its recent board meeting, concerns were raised by the members on the efficiency of the marketing campaign currently being used. The board decides to rope in analytics professional to improve the existing campaign.

### Data Description

• Age: The age of the individual in years.

• Gender: The gender of the individual, categorized as male or female.

• Profession: The occupation or profession of the individual.

• Marital_status: The marital status of the individual, such as married &, single

• Education: The educational qualification of the individual Graduate and Post Graduate

• No_of_Dependents: The number of dependents (e.g., children, elderly parents) that the individual supports financially.

• Personal_loan: A binary variable indicating whether the individual has taken a personal loan "Yes" or "No"

• House_loan: A binary variable indicating whether the individual has taken a housing loan "Yes" or "No"

• Partner_working: A binary variable indicating whether the individual's partner is employed "Yes" or "No"

• Salary: The individual's salary or income.

• Partner_salary: The salary or income of the individual's partner, if applicable.

• Total_salary: The total combined salary of the individual and their partner (if applicable).

• Price: The price of a product or service.

• Make: The type of automobile

```python
# as we can also see that there are spelling mistakes in the gender column which can be rectified  as followed 
df['Gender'] = df['Gender'].replace('Femal','Female')
df['Gender'] = df['Gender'].replace('Femle','Female')
```

```python
# mode of gender 
df['Gender'].mode()
```

```python
# impute null values of gender with mode
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
print('Gender\n',  df['Gender'].value_counts())
```

```python
# check 
df.isnull().sum()
```

```python
# imputing for partner salary

df.loc[df['Total_salary'] > df['Salary'], 'Partner_salary'] = df['Total_salary'] - df['Salary']
df.loc[df['Total_salary'] <= df['Salary'], 'Partner_salary'] = 0
```

Applied condition that, 
1. If ,  ‘Total_salary’  >  ‘Salary’  
then,  ‘Partner_salary’  =   ‘Total_salary’  -  ‘Salary’ 
2. If,  ‘Total_salary   ≯  ‘Salary’ 
Then,  ‘Partner_salary’  =  0

```python
numerical_cols = ['Salary', 'Age', 'Partner_salary', 'Total_salary', 'Price']
df_melted = df[numerical_cols].melt(var_name='Variable', value_name='Value')

# Setting plot style
plt.figure(figsize=(12, 6))
sns.boxplot(x='Variable', y='Value', data=df_melted, palette='Set1')  # Vibrant color palette

# Adding title and formatting
plt.title('Box Plot for Salary, Age, Partner_Salary, Total_Salary, and Price', fontsize=14)
plt.xlabel('')
plt.ylabel('Value', fontsize=12)
plt.xticks(rotation=30)

# Add figure caption below the plot
plt.figtext(0.5, -0.1, 'Numerical variables', wrap=True, 
            horizontalalignment='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
numeric_cols = ['Salary', 'Age', 'Partner_salary', 'Total_salary', 'Price']

# Loop over each numeric column and create a separate box plot
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color='#1f77b4')
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

**Univariate Analysis for Numeric Variables**

```python
numeric_cols = ['Salary', 'Age', 'Partner_salary', 'Total_salary', 'Price']

# Define the color used in your provided plots
plot_color = '#1f77b4'

# Loop over each numeric column
for col in numeric_cols:
    plt.figure(figsize=(12, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data=df, x=col, bins=20, color=plot_color, kde=False)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color=plot_color)
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)

    plt.tight_layout()
    plt.show()
```

**Univariate analysis of categorial**

```python
# List of categorical columns
cat_cols = [
    'Gender', 'Profession', 'Marital_status', 'Education',
    'Personal_loan', 'House_loan', 'No_of_Dependents',
    'Partner_working', 'Make'
]

# Define color palette
colors = ['#1f77b4', '#ff7f0e']  # Blue and Orange from previous image

# Plot in pairs
for i in range(0, len(cat_cols), 2):
    plt.figure(figsize=(12, 4))
    
    # First variable
    plt.subplot(1, 2, 1)
    sns.countplot(data=df, x=cat_cols[i], palette=colors)
    plt.title(f'{cat_cols[i]} Distribution')
    plt.xlabel(cat_cols[i])
    plt.ylabel('Count')
    
    # Second variable (if exists)
    if i + 1 < len(cat_cols):
        plt.subplot(1, 2, 2)
        sns.countplot(data=df, x=cat_cols[i + 1], palette=colors)
        plt.title(f'{cat_cols[i + 1]} Distribution')
        plt.xlabel(cat_cols[i + 1])
        plt.ylabel('Count')
    
    plt.tight_layout()
    plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Total_salary'].quantile(0.25)
Q3 = df['Total_salary'].quantile(0.75)

# Compute the IQR
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers
df_capped = df.copy()
df_capped['Total_salary'] = df_capped['Total_salary'].apply(
    lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x)
)

# Display summary
print("Original shape:", df.shape)
print("Shape after removing outliers:", df_capped.shape)
```
